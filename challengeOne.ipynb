```python
%%js
class Trie {
    constructor() {
        this.root = {};
    }

    insert(word) {
        let node = this.root;
        for (let char of word) {
            if (!node[char]) {
                node[char] = {};
            }
            node = node[char];
        }
        node.isEnd = true;
    }

    search(word) {
        let node = this.root;
        for (let char of word) {
            if (!node[char]) {
                return false;
            }
            node = node[char];
        }
        return node.isEnd === true;
    }

    startsWith(prefix) {
        let node = this.root;
        for (let char of prefix) {
            if (!node[char]) {
                return false;
            }
            node = node[char];
        }
        return true;
    }
}

//test
const operations = ["Trie", "insert", "search", "search", "startsWith", "insert", "search"];
const values = [[], ["apple"], ["apple"], ["app"], ["app"], ["app"], ["app"]];

let trie;
let output = [];

for (let i = 0; i < operations.length; i++) {
    const op = operations[i];
    const val = values[i];

    if (op === "Trie") {
        trie = new Trie();
        output.push(null);
    } else if (op === "insert") {
        trie.insert(val[0]);
        output.push(null);
    } else if (op === "search") {
        output.push(trie.search(val[0]));
    } else if (op === "startsWith") {
        output.push(trie.startsWith(val[0]));
    }
}

element.innerHTML = `
<pre>
Input: ${JSON.stringify(operations)}
       ${JSON.stringify(values)}

Output: ${JSON.stringify(output)}



</pre>`;
```
